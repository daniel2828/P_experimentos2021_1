## Tarea 1

Carga los datos de una forma relativamente automatizada.

Recuerda que en Python las funciones son elementos de primer nivel y se pueden guardar en listas si así lo deseas.

Crea una tabla que tenga 4 filas, una por dataset y tenga como columnas:
- Número de ejemplos.
- Número de atributos.
- Número de clases
- Ratio de desequilibrio (tamaño de la clase más pequeña / tamaño de la clase más grande)

```python
'''
Pista

Usar funciones en una lista, como si se tratase de cualquier otro tipo de datos.

'''
%matplotlib inline
#Creamos las 4 funciones para la carga de datos
from sklearn import datasets
def carga_datos1():
    return datasets.load_iris()
def carga_datos2():
    return datasets.load_digits()
def carga_datos3():
    return datasets.load_wine()
def carga_datos4():
    return datasets.load_breast_cancer()
#Creamos una lista para la carga de datos y otra para los nombres
lista_carga_datos = [carga_datos1, carga_datos2, carga_datos3,carga_datos4]
nombre_listas =[ "Iris", "Digits", "Wine",  "Cancer"]
i = 0
#Generamos todos los datos pedidos
for f in lista_carga_datos:
    print("##############################################################")
    print("Lista: " , nombre_listas[i])
    i=i+1
    datos = f()
    print("Número de ejemplos: " , len(datos.data))
    print("Número de atributos: " , len(datos.feature_names))
    print("Número de clases: " , len(datos.target_names))
    #Creamos
    countCl = {}
    for element in datos.target:
        if element in countCl:
            countCl[element] = countCl[element] + 1
        else:
            countCl[element] = 1
    all_values = countCl.values()
    max_value = max(all_values)
    min_value = min(all_values)
    print("Ratio de desequilibro: %0.2f" % (min_value/max_value))
    print("##############################################################")
```

## Tarea 2

- Realiza experimentos con una batería de 6 algoritmos, 1 sería Random Forest con 100 árboles y los otros 5 a tu elección.

- Usa validación cruzada de 5 folds.

- Obten el accuracy de cada método para cada conjunto de datos.

- Crea una tabla de 4 x 6, con el accuracy de cada método.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
import pandas as pd
#Definimos una función para cada algoritmo
def alg_1():
    return RandomForestClassifier(n_estimators = 100)
def alg_2():
    return DecisionTreeRegressor(max_depth=5)
def alg_3():
    return RandomForestClassifier(n_estimators = 50)
def alg_4():
    return KNeighborsRegressor(5)
def alg_5():
    return GradientBoostingRegressor(random_state=1)
def alg_6():
    return LinearRegression()
#Lista de algoritmos y lista de nombres
lista_carga_algoritmos = [alg_1, alg_2,alg_3,alg_4, alg_5, alg_6]
lista_nombre_algoritmos = ["Random Forest (100)" ,  "Decision Tree (5)",  "Random Forest(50)" ,
                           "Neighbors regresor(5)",  "Gradient Boosting(1)",  "Linear regression"]
i = 0
#Lista para meter todos los datos
lista =[]
for f in lista_carga_datos:
    
    print("##############################################################")
    print("Lista: " , nombre_listas[i])
    i=i+1
    datos = f()
    #Obtenemos los datos de la lista de carga de datos
    X = datos.data
    y = datos.target
    tplList = []
    #Recorremos los algoritmos
    for a in lista_carga_algoritmos:
        model = a()
        scores = cross_val_score(model, X, y, cv=5)
        print(scores.mean())
        #Añadimos a la lista de valores
        tplList.append(scores.mean())
    #Añadimos a la lista global la lista para esa fila
    lista.append(tplList)
dfObj = pd.DataFrame(lista, columns = lista_nombre_algoritmos, index=nombre_listas)
dfObj
```

## Tarea 3

Determina cual es el mejor algoritmo de manera *global*

Vamos a usar para ello el método del ranking promedio.

En ese método se ordenan los algoritmos de acuerdo a su desempeño y se les dan puntos según el lugar que ocupan, al algoritmo con mejor desempeño se le da 1 punto, al segundo mejor 2, si hay dos métodos empatados se reparten los puntos etc.

El ranking promedio es la media de los rankings obtenidos para cada conjunto de datos.

Calcula el ranking promedio de cada método.

```python
import pandas as pd

df_results = pd.DataFrame({"Método 1":[0.9,0.95,0.99,0.85],
              "Método 2":[0.91,0.94,0.9,0.85],
              "Método 3":[0.92,0.84,0.8,0.81]})
df_results
```
